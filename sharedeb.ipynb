```python
# 일단 EB normalizing
import pandas as pd
import numpy as np

ebs = []
group='02'
for i, subname in enumerate (['김예은','설혜경','임유나']):
    groupsub = group+'0'+str(i+1)
    DATADIR = f'/home/jiunchoi/OFD/OFD_BHV_clean/group-{group}/'
    day1fname = f'group-{group}_day1_posthoc_{subname}님.xlsx'
    day2fname = f'sub-{groupsub}_day2_posthoc.xlsx'
    data = pd.read_excel(DATADIR+day1fname)
    eb = data['segmentation']
    ebs.append(np.array(eb))

# 결측값 보간 함수
def fill_missing(array):
    for i in range (1, len(array)):
        if array[i] > 0:
            prev = array[i-1]
        else:
            array[i] = prev
    return array

one = []
for i in range (3):
    ebs[i] = fill_missing(ebs[i])
    one.append(ebs[i][:-1]-ebs[i][1:])

sharedeb = np.zeros_like(ebs[0])

for i in range (len(ebs[0])-3):
    tmp = np.vstack((one[0][i:i+3],one[1][i:i+3],one[2][i:i+3])).T
    #print('i', i,'\n', tmp)
    if np.nansum(tmp) < -1:
        if np.nansum(tmp[0]) == -3:
            sharedeb[i] = 3 ##
        elif np.nansum(tmp[0]) == -2:
            sharedeb[i] = 2 ##
        else:
            x=np.where(tmp!=0)[0]
            med = (np.median(x))
            if med%1 == 0.5:
                med = med-0.5
            sharedeb[i+int(med)] = 1
    elif np.nansum(tmp) == -1:
        if np.nansum(tmp[0])==-1 and i>0 and sharedeb[i-1] == 0:
            sharedeb[i] = 10 ##
```

```python
sharedebeb = np.zeros_like(sharedeb)
evno = 1
for i in range (len(sharedeb)):
    sharedebeb[i] = evno #얘의 위치에 따라 한칸씩 먼저 찍히나(sharedeb) 한칸씩 나중에 찍히나 결정
    if sharedeb[i] != 0:
        evno += 1
    # print(sharedebeb[i])
```

```python
ebb = np.vstack((sharedebeb, ebs[0], ebs[1], ebs[2])).T

for i in range (len(ebb)):
    print(ebb[i])

# 할 거!!! 일단 엑셀을 만들어!! sent랑 sharedeb!! 그걸보고 RSA용 노가다를 해!!! 그리고 recall데이터가지고 좀 놀아봐야해
group = '02'; subname='김예은'
DATADIR = f'/home/jiunchoi/OFD/OFD_BHV_clean/group-{group}/'
day1fname = f'group-{group}_day1_posthoc_{subname}님.xlsx'
data = pd.read_excel(DATADIR+day1fname)
ebb = pd.DataFrame(ebb)

datacon = pd.concat([data['speaker'], data['sentence'], ebb], axis=1)
datacon
```
